# Curve for lr=$10^{-1}$

This notebook display the evolution of the loss function of different methods.

## Import

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Switcher

```python
class CSV():
    def __init__(self):
        self.switch = {
            "static" : None,
            "fastai" : None
        }
    
    def set(self, name:str):
        nom = name.split('.')
        if nom[1]=='csv':
            self.switch[nom[0]] = pd.read_csv(name)
```

## Reading

```python
path = os.path.join("/home/julien/traineeship/curve/e-1")
csv = CSV()
csv_name = os.listdir(path)

for name in csv_name:
    csv.set(name)
    
list_name = ["fastai", "static", "one_cycle"]
```

## Plot

```python
plt.figure()
for lab in list_name:
    plt.plot(csv.switch.get(lab)['Step'], csv.switch.get(lab)['Value'], label=lab)
plt.title("Curve of loss function")
plt.xlabel("Iterations")
plt.ylabel("Loss Function")
plt.legend()
plt.show()
```
