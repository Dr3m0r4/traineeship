```python
rc('text', usetex=True)

alpha = ['alpha', 'beta', 'gamma','delta','epsilon','dzeta','eta','theta','iota','kappa','lambda','mu','nu','ksi','omicron','pi','rho','final_sigma','sigma','tau','upsilon','phi','khi','psy','omega']
gamma = []
for lt in range(0x03b1,0x03ca):
    gamma.append(chr(lt))

Greek = {}
for i in range(len(alpha)):
    Greek[alpha[i]] = gamma[i]
print(Greek)
```

```python
maxi = eval(input('please enter the initial learning rate : '))
theta = 2 #float(input('please enter the period of sinusoid function : '))
print("the period of the sinusoid is {}".format(theta))
#beta = float(input('please enter the reduction of the function : '))
print('the coefficent of reduction of the function is between 20 and 20000')
space = 2*np.logspace(1, 3, 30)
f = lambda x, y, beta: maxi*(np.cos(x/theta*np.pi+y*np.pi)+1)/(x/beta+1)
```

```python
X = range(501)
k = 0

total = []

for b in space :
    Y = []
    for x in X:
        if x>0 and f(x-1,k,b)<1e-7:
            k+=1
        Y.append(f(x,k,b))

    X = np.array(X)
    Y = np.array(Y)
    total.append(Y)
total = np.array(total)
```

## Figure

```python
for i in range(space.shape[0]) :
    plt.figure(i)
    plt.title("Modification of the Learning Rate\nwith reduction coefficient of {}".format(space[i]))
    plt.plot(X, total[i])
    plt.xlabel("Iterations")
    plt.ylabel("Learning Rate")
    plt.show()
```
