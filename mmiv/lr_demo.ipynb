## FastAI Version
#### (Version 9)

We know that Fastai has already some algorithms which permits to do some superconvergence. Now we will see how looks like their one cycle algorithm.

### Import

```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

### Definition

```python
def is_tuple(var):
    return isinstance(var,tuple)

def is_list(var):
    return isinstance(var, list)

def annealing_cos(start, end, pct:float):
    cos_out = np.cos(np.pi*pct)+1
    return end + (start-end)/2 * cos_out

class Scheduler():
    def __init__(self, vals, n_iter:int):
        self.start, self.end = (vals[0],vals[1]) if is_tuple(vals) else (vals,0)
        self.n_iter = max(1,n_iter)
        self.func = annealing_cos
        self.n = 0
    
    def restart(self):
        self.n = 0
    
    def step(self):
        self.n += 1
        return self.func(self.start, self.end, self.n/self.n_iter)
    
    @property
    def is_done(self)->bool:
        return self.n >= self.n_iter
    
def steps(dico):
    return [Scheduler(step, n_iter) for (step, n_iter) in zip(dico['steps_cfg'], dico['phases'])]

def rule(train, lr_scheds, idx_s):
    if train :
        if idx_s >= len(lr_scheds):
            return {'stop' : True}
        lr = lr_scheds[idx_s].step()
        if lr_scheds[idx_s].is_done:
            idx_s +=1
        return {'lr' : lr, 'idx' : idx_s}
```

### Function

```python
train = True
n = int(input("Please enter the number of iterations : "))

lr_max = eval(input("Please enter the max learning rate :"))
div_factor = 7 #-np.log10(lr_max)/10 if lr_max < 1 else np.log10(lr_max)*3 # int(input("Please enter the value of reduction : "))

if is_list(lr_max):
    lr_max = np.array(lr_max)
pct = 1/lr_max if lr_max > 3 else 0.3
a = int(n*pct)
b = n-a

phases = (a,b)
final_div = div_factor * 1e3
low_lr = lr_max/div_factor
lr_min = lr_max/final_div
steps_cfg = ((low_lr, lr_max),(lr_max,lr_min))

X,Y = np.linspace(0,n,n+1),[]

lr_prop = steps({'steps_cfg':steps_cfg, 'phases':phases})

Y.append(lr_prop[0].start)
res = {}
while not res.get("stop",False):
    res = rule(train, lr_prop, res.get('idx', 0))
    Y.append(res.get("lr",0))

Y.pop(-1)
Y = np.transpose(np.array(Y))
```

### Figure

```python
plt.figure()
print(X.shape, Y.shape)
if X.shape == Y.shape:
    plt.plot(X,Y)
else:
    for i in range(Y.shape[0]):
        plt.plot(X,Y[i])
plt.title("Modification of the Learning Rate")
plt.xlabel("Iterations")
plt.ylabel("Learning Rate")
plt.savefig(os.path.join('report','curve.eps'))
```

But the higher learning rate have a better results with this 9th version than with the 7th one.
