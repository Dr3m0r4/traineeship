## Code

```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
maxi = eval(input('please enter the initial learning rate : '))
theta = float(input('please enter the period of sinusoid function : '))
beta = float(input('please enter the reduction of the function : '))
f = lambda x, y: maxi*(np.cos(x/theta*np.pi+y*np.pi)+1)/(x/beta+1)
```

```python
X = range(501)
k = 0
Y = []

for x in X:
    if x>0 and f(x-1,k)<1e-7:
        k+=1
    Y.append(f(x,k))
```

```python
X = np.array(X)
Y = np.array(Y)
```

## Figure

```python
plt.plot(X, Y)
plt.show()
```
